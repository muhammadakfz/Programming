```python
import pandas as pd
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('air_polution.csv')
df
```

```python
# Sort the DataFrame by AQI values in descending order
df_sorted = df.sort_values(by='AQI Value', ascending=False)

# Initialize variables
top_countries = []
top_5_countries = []

# Iterate through the sorted DataFrame
for index, row in df_sorted.iterrows():
    country = row['Country']
    if country not in top_countries:
        top_countries.append(country)
        top_5_countries.append(row)
    
    # If we have collected data for 5 different countries, exit the loop
    if len(top_5_countries) == 5:
        break

# Create a countplot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.barplot(y="AQI Value", x="Country", data=pd.DataFrame(top_5_countries), palette="ch:start=.2,rot=-.3")

# Customize the plot
plt.title("Top 5 Countries with Highest Average AQI Values")
plt.ylabel("AQI Value")
plt.xlabel("Country")

# Show the plot
plt.show()
```

```python
import sklearn.linear_model as linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.scatterplot(data=df, x='AQI Value', y='PM2.5 AQI Value')
```

```python
aqi_value = df['AQI Value'].values.reshape(-1, 1)
lign_reg = linear_model.LinearRegression()
lign_reg.fit(aqi_value, df['PM2.5 AQI Value'])
```

```python
# Plot the regression line
plt.scatter(aqi_value, df['PM2.5 AQI Value'])
plt.plot(aqi_value, lign_reg.predict(aqi_value), color='red')
plt.xlabel('AQI Value')
plt.ylabel('PM2.5 AQI Value')
plt.show()
```
